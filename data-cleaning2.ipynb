```python
import pandas as pd
import numpy as np
import matplotlib as mlt
%matplotlib inline
```

```python
data = pd.read_csv('cleaned_1.csv')
data.head()
```

```python
data = data.drop(data.columns[[0,1]],axis = 1)
data.head()
```

```python
#data['date_time'] = pd.to_datetime(data['date_time'], format = '%Y-%m-%d %H:%M:%S')
#data.dtypes
```

```python
temp = pd.DatetimeIndex(data['date_time'])
data.insert(loc=1,column='search_year',value=temp.year)
data.insert(loc=2,column='search_month',value=temp.month)
data.insert(loc=3,column='search_day',value=temp.day)
data.insert(loc=4,column='search_dayofweek',value=temp.dayofweek)
data.head()
```

```python
data.groupby('site_id').count()
```

```python
data.groupby('prop_id').count()
```

```python
# Group values for Series and add 'cnt' column with count
data['hotel_cnt'] = data.groupby(['prop_id'])['date_time'].transform('count')
# Drop indexes for count value < 100, and dropping 'cnt' column
new_data = data.drop(data[data.hotel_cnt < 100].index)[data.columns]
```
