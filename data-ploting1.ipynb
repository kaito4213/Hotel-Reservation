```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('cleaned_2.csv')
data.head()
```

```python
data.columns
```

```python
data = data.drop(data.columns[0], axis =1)
data.head()
```

```python
data['date_time'].dtype
```

```python
data['checkin_date'] = pd.to_datetime(data.date_time) + pd.to_timedelta(pd.np.ceil(data.srch_booking_window), unit="D")
data.head()
```

```python
data.head(50)
```

```python
#delete time in column date_time and checkin_date
data['booking_date']= pd.to_datetime(data['date_time'])
data['short_booking_date'] = data['booking_date'].apply(lambda x : x.date().strftime('%y-%m-%d'))
```

```python
data.head()
```

```python
data['short_checkin_date'] = data['checkin_date'].apply(lambda x : x.date().strftime('%y-%m-%d'))
data.head()
```

```python
count = data.groupby('prop_id').size().reset_index(name='counts')
count.idxmax()
```

```python
count.iloc[[463]]
```

```python
test = data.loc[data['prop_id'] == 104517]
test
```

```python
price = test.groupby(['short_booking_date', 'short_checkin_date'])['price_usd'].mean().reset_index()
price
```

```python
price.sort_values(['short_booking_date', 'short_checkin_date'], ascending=[True, True])
price
```

```python
price.plot(x='short_checkin_date', y='price_usd')
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

price_data = remove_outlier(price, 'price_usd')
price_data.plot(x='short_checkin_date', y='price_usd')
```

```python
price_data = price_data.loc[(price_data['price_usd'] > 25)]
price_data.plot(x='short_checkin_date', y='price_usd')
```

```python
pt = price_data.pivot_table(index='short_booking_date', columns='short_checkin_date', values='price_usd')
pt.head()
```

```python
f, ax = plt.subplots(figsize = (20, 8))
sns.heatmap(pt, linewidths = 0, ax = ax, cmap = 'BuPu')
ax.set_title('price trends')
ax.set_xlabel('short_checkin_date')
ax.set_ylabel('short_booking_date')

f.savefig('sns_heatmap_normal.png', bbox_inches='tight')
```
